#### Receipts table

```python
def date_transfer(milliseconds):
    if milliseconds is None:
        return None
    return datetime.datetime.fromtimestamp(milliseconds/1000.0).strftime("%Y-%m-%d")
```

```python
receipts = []
with open('data/receipts.json') as f:
    for line in f:
        receipts.append(json.loads(line))

rcp_id_list = []
bonusPointsEarned_list = []
createDate_list = []
dateScanned_list = []
finishedDate_list = []
modifyDate_list = []
pointsAwardedDate_list = []
pointsEarned_list = []
purchaseDate_list = []
purchasedItemCount_list = []
barcode_list = []
description_list = []
finalPrice_list = []
itemPrice_list = []
quantityPurchased_list = []
rewardsReceiptStatus_list = []
totalSpent_list = []
userId_list = []

for i in range(len(receipts)):
    rcp = receipts[i]
    if rcp.get('rewardsReceiptItemList', None):
        for j in range(len(rcp['rewardsReceiptItemList'])):
            rcp_id = rcp['_id']['$oid']
            bonusPointsEarned = rcp.get('bonusPointsEarned', None)
            createDate = date_transfer(rcp.get('createDate').get('$date', None))
            dateScanned = date_transfer(rcp.get('dateScanned').get('$date', None))
            finishedDate = date_transfer(rcp.get('finishedDate', None).get('$date', None) if rcp.get('finishedDate', None) else None)
            modifyDate = date_transfer(rcp.get('modifyDate').get('$date', None))
            pointsAwardedDate = date_transfer(rcp.get('pointsAwardedDate').get('$date', None) if rcp.get('pointsAwardedDate', None) else None)
            pointsEarned = rcp.get('pointsEarned', None)
            purchaseDate = date_transfer(rcp.get('purchaseDate').get('$date', None) if rcp.get('purchaseDate', None) else None)
            purchasedItemCount = rcp.get('purchasedItemCount', None)
            barcode = rcp['rewardsReceiptItemList'][j].get('barcode', None)
            description = rcp['rewardsReceiptItemList'][j].get('description', None)
            finalPrice = rcp['rewardsReceiptItemList'][j].get('finalPrice', None)
            itemPrice = rcp['rewardsReceiptItemList'][j].get('itemPrice', None)
            quantityPurchased = rcp['rewardsReceiptItemList'][j].get('quantityPurchased', None)
            rewardsReceiptStatus = rcp.get('rewardsReceiptStatus', None)
            totalSpent = rcp.get('totalSpent', None)
            userId = rcp.get('userId', None)
            rcp_id_list.append(rcp_id)
            bonusPointsEarned_list.append(bonusPointsEarned)
            createDate_list.append(createDate)
            dateScanned_list.append(dateScanned)
            finishedDate_list.append(finishedDate)
            modifyDate_list.append(modifyDate)
            pointsAwardedDate_list.append(pointsAwardedDate)
            pointsEarned_list.append(pointsEarned)
            purchaseDate_list.append(purchaseDate)
            purchasedItemCount_list.append(purchasedItemCount)
            barcode_list.append(barcode)
            description_list.append(description)
            finalPrice_list.append(finalPrice)
            itemPrice_list.append(itemPrice)
            quantityPurchased_list.append(quantityPurchased)
            rewardsReceiptStatus_list.append(rewardsReceiptStatus)
            totalSpent_list.append(totalSpent)
            userId_list.append(userId) 

    else:
        rcp_id = rcp['_id']['$oid']
        bonusPointsEarned = rcp.get('bonusPointsEarned', None)
        createDate = date_transfer(rcp.get('createDate').get('$date', None))
        dateScanned = date_transfer(rcp.get('dateScanned').get('$date', None))
        finishedDate = date_transfer(rcp.get('finishedDate', None).get('$date', None) if rcp.get('finishedDate', None) else None)
        modifyDate = date_transfer(rcp.get('modifyDate').get('$date', None))
        pointsAwardedDate = date_transfer(rcp.get('pointsAwardedDate').get('$date', None) if rcp.get('pointsAwardedDate', None) else None)
        pointsEarned = rcp.get('pointsEarned', None)
        purchaseDate = date_transfer(rcp.get('purchaseDate').get('$date', None) if rcp.get('purchaseDate', None) else None)
        purchasedItemCount = rcp.get('purchasedItemCount', None)
        barcode = None
        description = None
        finalPrice = None
        itemPrice = None
        quantityPurchased = None
        rewardsReceiptStatus = rcp.get('rewardsReceiptStatus', None)
        totalSpent = rcp.get('totalSpent', None)
        userId = rcp.get('userId', None)        

        rcp_id_list.append(rcp_id)
        bonusPointsEarned_list.append(bonusPointsEarned)
        createDate_list.append(createDate)
        dateScanned_list.append(dateScanned)
        finishedDate_list.append(finishedDate)
        modifyDate_list.append(modifyDate)
        pointsAwardedDate_list.append(pointsAwardedDate)
        pointsEarned_list.append(pointsEarned)
        purchaseDate_list.append(purchaseDate)
        purchasedItemCount_list.append(purchasedItemCount)
        barcode_list.append(barcode)
        description_list.append(description)
        finalPrice_list.append(finalPrice)
        itemPrice_list.append(itemPrice)
        quantityPurchased_list.append(quantityPurchased)
        rewardsReceiptStatus_list.append(rewardsReceiptStatus)
        totalSpent_list.append(totalSpent)
        userId_list.append(userId)   
        
receipts_df = pd.DataFrame({'rcp_id': rcp_id_list})

column_nm ={'bonusPointsEarned':bonusPointsEarned_list, 'createDate':createDate_list, 
            'dateScanned':dateScanned_list, 'finishedDate':finishedDate_list,
            'modifyDate':modifyDate_list, 'pointsAwardedDate':pointsAwardedDate_list, 
            'pointsEarned':pointsEarned_list, 'purchaseDate':purchaseDate_list, 
            'purchasedItemCount':purchasedItemCount_list, 'barcode':barcode_list, 
            'description':description_list, 'finalPrice':finalPrice_list, 'itemPrice':itemPrice_list,
            'quantityPurchased':quantityPurchased_list, 'rewardsReceiptStatus':rewardsReceiptStatus_list, 
            'totalSpent':totalSpent_list, 'userId':userId_list}


for col_nm, col in column_nm.items():
    receipts_df[col_nm] = col
```
